Below function will read the data from the Streaming_Linux_process.csv and create list with data frame

```python
def readCSVFile(fileName):
    list=[] # empty list
    with open('Streaming_Linux_process.csv', 'rt') as f: # open the file
        reader = csv.DictReader(f) # read in dictionary format
        # Interate each rows with datatype and add to the list
        for row in reader: 
            list.append({'sequence':int(row['sequence']), 
                         'machine':int(row['machine']),
                         'PID':int(row['PID']),
                         'TRUN':int(row['TRUN']),
                         'TSLPI':int(row['TSLPI']),
                         'TSLPU':int(row['TSLPU']),
                         'POLI':str(row['POLI']),
                         'NICE':int(row['NICE']),
                         'PRI':int(row['PRI']),
                         'RTPR':int(row['RTPR']),
                         'CPUNR':int(row['CPUNR']),
                         'Status':str(row['Status']),
                         'EXC':int(row['EXC']),
                         'State':str(row['State']),
                         'CPU':float(row['CPU']),
                         'CMD':str(row['CMD'])})
#             list.append(row)
    return list
```

Below function will sent the topic and data to the kafka.

```python
def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data) # sending the data to kafka
        print('Message published successfully. Data: ' + str(data)) 
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
```

Below function will connect to the kafka producer for an ip

```python
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'], # connect to the ip
                                  value_serializer=lambda x: dumps(x).encode('ascii'), # convert the datatype to ascii
                                  api_version=(0, 10)) # version
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
```
