Below function will read the data from the Streaming_Linux_memory.csv and create list with data frame

```python
def readCSVFile(fileName):
    list=[]
    with open('Streaming_Linux_memory.csv', 'rt') as f:
        reader = csv.DictReader(f)
        for row in reader:
            list.append({'sequence':int(row['sequence']),
                         'machine':int(row['machine']),
                         'PID':int(row['PID']),
                         'MINFLT':str(row['MINFLT']),
                         'MAJFLT':str(row['MAJFLT']),
                         'VSTEXT':str(row['VSTEXT']),
                         'VSIZE':str(row['VSIZE']),
                         'RSIZE':str(row['RSIZE']),
                         'VGROW':str(row['VGROW']),
                         'RGROW':str(row['RGROW']),
                         'MEM':float(row['MEM']),
                         'CMD':str(row['CMD'])
                        })
#             list.append(row)
    return list
```

Below function will sent the topc and data to the kafka.

```python
def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)# sending the data to kafka
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
```

Below function will connect to the kafka producer for an ip

```python
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'], # connect to the ip
                                  value_serializer=lambda x: dumps(x).encode('ascii'),# convert the datatype to ascii
                                  api_version=(0, 10))# version
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
```
